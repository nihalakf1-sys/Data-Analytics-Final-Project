## 2: Data Cleaning & Pre-processing
Data cleaning and preprocessing are important steps in data analytics. Raw datasets often contain missing values, duplicate records, and inconsistent formats. Before performing analysis, the dataset must be cleaned and prepared to ensure accurate results.

For this project, the New York Real Estate Data 2026 dataset was cleaned and preprocessed using Python and the Pandas library.

### 1. Handling Missing Data

Missing values were checked using the isnull() function. Missing data can affect the accuracy of analysis, so they were handled appropriately.

Some missing values were filled using suitable methods such as forward filling or replacing them with average values.

```python
# Check Missing Values
print(df.isnull().sum())
```

### Handling Missing Values
Missing values in numerical columns such as sqft, beds, baths were handled using median values to reduce the impact of outliers.
Categorical column sub_type was filled using the mode (most frequent value).
The garage column was filled with 0 assuming absence of garage.
The stories column was filled using mode as it represents discrete values.

```python
df.fillna({
    'sub_type': 'Unknown',
    'sqft': df['sqft'].median(),
    'stories': df['stories'].mode()[0],
    'beds': df['beds'].median(),
    'baths': df['baths'].median(),
    'baths_full': df['baths_full'].median(),
    'baths_full_calc': df['baths_full_calc'].median(),
    'garage': 0
}, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Check duplicates
print("Duplicate rows:", df.duplicated().sum())
```

### 2. Removing Duplicate Records

Duplicate rows in the dataset may cause incorrect analysis results. These duplicates were identified and removed using the drop_duplicates() function.

### 🔗2. Bivariate Analysis

Bivariate analysis examines the relationship between two variables.
